```python
import pandas as pd
from sklearn.model_selection import train_test_split
import math
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import metrics
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pycon-2016-tutorial/master/data/sms.tsv',header=None,delimiter = '\t') 
df
```

```python
 X=df[1]
y=df[0]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
corpus=X_train
vectorizer = CountVectorizer()
x = vectorizer.fit_transform(corpus)
col=vectorizer.get_feature_names()
trainData=pd.DataFrame(x.toarray(),columns=col,index=y_train.index)
print(trainData.shape[0])
trainData['output']=y_train
print(trainData['output'].value_counts())
spam=(trainData[trainData['output']=='spam'])
ham=(trainData[trainData['output']=='ham'])
probSpam=spam.shape[0]/y_train.shape[0]
probHam=1-probSpam
del trainData['output']
del spam['output']
del ham['output']
totalSpamWords=(spam.sum()).sum()
totalHamWords=(ham.sum()).sum()
```

```python
trainingDict=dict([(key,dict([(keys, 0) for keys in df[0].value_counts().keys()])) for key in vectorizer.get_feature_names()])
trainingDict
```

```python
spamSeries=spam.sum()
ind=spamSeries.index
for i in ind:
    trainingDict[i]['spam']=spamSeries[i]/totalSpamWords
hamSeries=ham.sum()
ind=hamSeries.index
for i in ind:
    trainingDict[i]['ham']=hamSeries[i]/totalHamWords
trainingDict
```

```python
def prediction(sent):
    sent=sent.lower().split()
    s=1
    h=1
    for i in sent:
        try:
            s=s*trainingDict[i]['spam']
            h=h*trainingDict[i]['ham']
        except KeyError:
            pass
    return "spam" if s*probSpam>h*probHam else "ham"
```

```python
y_test.index=range(0,1672)
size=X_test.shape[0]
predict=[]
for i in range(0,size):
    sentence=X_test.iloc[i]
    predict.append(prediction(sentence))
metrics.accuracy_score(y_test, predict)*100
```

```python
print("Enter the sentence to be classified as a Spam or ham")
sentence=input()
print(prediction(sentence))
```
